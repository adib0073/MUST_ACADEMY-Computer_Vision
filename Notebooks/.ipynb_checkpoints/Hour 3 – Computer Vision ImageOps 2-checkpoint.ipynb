### 1. Convolution vs Correlation? 

With the previous filter operations that we have tried so far, did we do convolution or correlation? What is the main difference?

![Convolution VS Correlation](dataset/ConvVSCorr.png)

The basic idea in correlation filtering:
1. Slide the center of the correlation kernel on the image
2. Multiply each weight in the correlation kernel by the pixel in the image
3. Sum these products

```python
# Utility function to display images
def show_image(img, fig_size = None):
    print(img.shape)
    img = img.astype('uint8')
    
    if fig_size is not None:
        plt.figure(figsize=fig_size)
    
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
    
    
src = cv2.imread('dataset/bicycle.png', 0)

# prepare averaging filter - example of correlation filter
kernel = np.ones((7, 7), dtype=np.float32)
kernel /= 49

# apply kernel to the original image
dst = cv2.filter2D(src, -1, kernel)

# concatenate images horizontally
result = np.concatenate((src, dst), axis=1)
show_image(result, (14,8))
```

![Correlation Filtering](dataset/CorrExplain.png)

Like correlation filtering, convolution filtering is also a linear filtering and it is more common then correlation filtering. But there is a small difference between correlation and convolution :

1. Flip the filter in both dimensions (bottom to top, right to left)
2. Apply correlation filtering

![Convolution Filtering - Filter flipped in both directions](dataset/ConvFilter1.png)

```python
from scipy import ndimage

# read image
src = cv2.imread('dataset/bicycle.png',0)

# prepare the filter
kernel = np.array([[1,1,1],[1,1,0],[1,0,0]])

# apply kernel to the original image using convolution filtering
dst_conv  = ndimage.convolve(src, kernel, mode='constant', cval=1.0)

# apply kernel to the original image using correlation filtering
dst_corr = cv2.filter2D(src, -1, kernel)

# concatenate images horizontally
result = np.concatenate((src, dst_corr, dst_conv), axis=1)
show_image(result, (15,8))
```

#### Is convolution filter used instead of correlation filter in CNN?

No, ironically, Convolutional Neural Network actually uses Cross-Correlation operation and hence correlational filtering is applied instead of convolutional  filtering. The very initial inception of CNNs is considered to be originated from the paper [Neocognitron](https://www.rctn.org/bruno/public/papers/Fukushima1980.pdf), which is known to use convolutional filtering, but later implementations of CNNs and most ML/DL packages are known to use Correlation Filtering instead of Convolutional filtering.

### 2. Image Binarization

Image binarization or image thresholding is extremely useful when for various computer vision application. This technique usually converts an image with varying pixel intensity values to a binary image, which will have only two possible image pixel values ... one representing the black pixel and one representing the white pixel. Hence image binarization produces a black and white image.

There are multiple types of image binarization techniques, we are going to discuss the following types:

1. Global Thresholding

2. Local Thresholding
    
    a. Adaptive Thresholding
    
    b. Otsu Thresholding

```python
# Global thresholding using open cv - only one threshold value set for entire image
shapesImage = cv2.imread('dataset/square.png', 0) # Using shapes image - has to be opened in gray scale
show_image(shapesImage)
```

```python
threshold_types = [cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV, cv2.THRESH_TRUNC, cv2.THRESH_TOZERO, cv2.THRESH_TOZERO_INV]
threshold_amount = 230

for th_type in threshold_types:
    _, binary_image = cv2.threshold(shapesImage,threshold_amount,255,th_type)
    show_image(binary_image)
```

```python
# Local thresholding is required when one global threshold value is not sufficient and every small region of interest should have
# a thershold value
```

```python
test_threshold = cv2.imread('dataset/threshold.png', 0) # Using shapes image - has to be opened in gray scale
show_image(test_threshold)
```

```python
# What happens if we apply Global Threshold
test_threshold = cv2.medianBlur(test_threshold,5)
_, th1 = cv2.threshold(test_threshold,127,255,cv2.THRESH_BINARY)
show_image(th1)
```

```python
th2 = cv2.adaptiveThreshold(test_threshold,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
show_image(th2)
th3 = cv2.adaptiveThreshold(test_threshold,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
show_image(th3)
```

```python
# Otsu thresholding
# Doesn't require to choose a filter value or a constant value and calculates this automatically by OTSU method
_,th4 = cv2.threshold(test_threshold,200,255,cv2.THRESH_OTSU)
show_image(th4)

# But usually it works well only for Bimodal images 
# Images with only two distinct image values, where the histogram would only consist of two peaks. 
# ideally,a good threshold would be in the middle of those two values. 
# Similarly, Otsu's method determines an optimal global threshold value from the image histogram.
```

### 3. Image Morphology

Morphological transformations on images is used on binary images to restructure a spotted or smudged binary image to produce a fine image. Very popular when used in manufacturing quality inspection process, to separate defect regions from equipment parts.

Popular morphological opeartions are:
1. Erosion
2. Dilation
3. Opening
4. Closing
5. Morphological Gradient
6. Morphological Top Hat 
7. Morphological Black Hat

```python
morphImage = cv2.imread('dataset/morphology.png',0)
show_image(morphImage)
```

```python
# Erosion
kernel = np.ones((10,10),np.uint8)
erosion = cv2.erode(morphImage,kernel,iterations = 6)
show_image(erosion)
```

```python
# Dilation
kernel = np.ones((10,10),np.uint8)
dilate = cv2.dilate(morphImage,kernel,iterations = 6)
show_image(dilate)
```

```python
morphImageOpen = cv2.imread('dataset/morphologyOpen.png',0)
show_image(morphImageOpen)

# Opening - erosion followed by dilation
kernel = np.ones((10,10),np.uint8)
opening = cv2.morphologyEx(morphImageOpen, cv2.MORPH_OPEN, kernel)
show_image(opening)
```

```python
morphImageClose = cv2.imread('dataset/morphologyClose.png',0)
show_image(morphImageClose)

# Closing - Dilation followed by Erosion
kernel = np.ones((10,10),np.uint8)
closing = cv2.morphologyEx(morphImageClose, cv2.MORPH_CLOSE, kernel)
show_image(closing)
```

```python
# Gradient
kernel = np.ones((10,10),np.uint8)
gradient = cv2.morphologyEx(morphImage, cv2.MORPH_GRADIENT, kernel)
show_image(gradient)
```

```python
# Top Hat and Black Hat
kernel = np.ones((10,10),np.uint8)
top_hat = cv2.morphologyEx(morphImage, cv2.MORPH_TOPHAT, kernel)
black_hat = cv2.morphologyEx(morphImage, cv2.MORPH_BLACKHAT, kernel)
show_image(top_hat)
show_image(black_hat)
```

### Image Boundary Issues

Let's talk about Valid and Same Convolution first, to explain some image boundary issues that can happen.

![Same and Valid Convolution](dataset/convolution_same_valid.PNG)

```python
leavesImage = cv2.imread('dataset/leaves.png')
show_image(leavesImage)
```

```python
filter_size = 11; filter_sigma = 2;
filter_kernel = cv2.getGaussianKernel(filter_size, filter_sigma)
filter_kernel = filter_kernel * filter_kernel.T

# Apply the kernel using different edge parameters. 
# OpenCV has even more options, please explore cv2.filter2D documentation for more
edge_params = {
               "constant": cv2.BORDER_CONSTANT, 
               "copy": cv2.BORDER_REFLECT,
               "mirror": cv2.BORDER_WRAP
              }

def filter_with_border_options(img, method = "copy"):
    temp_img = cv2.copyMakeBorder(img, filter_size, filter_size, filter_size, filter_size,edge_params[method])
        
    show_image(temp_img)
```

```python
# Replicate Border edges 
print('constant padded border')
filter_with_border_options(leavesImage, "constant")

# With Symmetric Border edges 
print('symmetric border')
filter_with_border_options(leavesImage, "copy")

# With Circular Replicate Border edges 
print('mirror opposite side borders')
filter_with_border_options(leavesImage, "mirror")
```
